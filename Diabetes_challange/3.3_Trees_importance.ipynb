```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd


data = pd.read_csv('Data/train_all_dummies.csv')
data.drop('Unnamed: 0', axis=1, inplace=True)

with open('models/minmax_RFE_Forest.pkl'):
    model = pickle.load(open('models/minmax_RFE_Forest.pkl', 'rb'))

forest = model.named_steps['classifier']
selector = model.named_steps['features_selector']
selected_features = selector.get_support(indices=True)
importances = forest.feature_importances_


plt.barh( data.columns[selected_features], importances)
```
