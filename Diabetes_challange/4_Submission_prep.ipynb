```python
import pandas as pd
from sklearn.metrics import roc_auc_score

train = pd.read_csv('Data/chosen_data.csv')
train.drop('Unnamed: 0', axis=1, inplace=True)
train_x, train_y  = train.drop('diagnosed_diabetes', axis=1), train['diagnosed_diabetes']

data_kaggle = pd.read_csv('Data/chosen_test.csv')
# data_kaggle.drop('id', axis=1, inplace=True)
data_kaggle.drop('Unnamed: 0', axis=1, inplace=True)

train_x
```

## Model

```python
import pickle


model = pickle.load(open('models/RF_best_features.pkl', 'rb'))

model.fit(train_x, train_y)
```

# Output csv


```python
import numpy as np

submission_name = "RF_pruned.csv"

ids = np.arange(700000, 1000000)

# data_kaggle.drop('id', axis=1, inplace=True)
probs = model.predict_proba(data_kaggle)[:, 1]
submission = pd.DataFrame({
    "id": ids,
    "diagnosed_diabetes": probs
})

submission
```

```python
import os
PATH = os.path.join('results',submission_name)

submission.to_csv(PATH, index=False, float_format="%.6f")
submission.head()
```
