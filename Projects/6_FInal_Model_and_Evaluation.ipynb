# Best Model fitted with all data

## Data

```python
import pandas as pd

data = pd.read_csv('new_Exam_Score_Prediction.csv')

y = data['exam_score']
X = data.drop('exam_score', axis=1)
print(X)
```

```python
import pickle
import os

with open(os.path.join('pipes', 'Linear.pkl'), 'rb') as f:
    Pipe = pickle.load(f)

Pipe.fit(X,y)
```
