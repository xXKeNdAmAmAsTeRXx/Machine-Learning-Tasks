# Coefficients Analysis of linear Regression
Analyzing the best model.

```python
# imports
import os
import pickle

import pandas as pd
from sklearn.model_selection import train_test_split
```

## Getting Data

```python
data = pd.read_csv('new_Exam_Score_Prediction.csv')

y = data['exam_score']
X = data.drop('exam_score', axis=1)

X
```

# Getting Model

```python
with open(os.path.join('pipes', 'Linear.pkl'), 'rb') as f:
    Pipe = pickle.load(f)

Pipe
```

# T-test for Regression Coefficients
Using the whole dataset.

```python
import numpy as np
import statsmodels.api as sm

names = X.columns
selector = Pipe.named_steps['features_selector']
mask = selector.get_support()
selected_names = np.array(names)[mask]

test_x = X[selected_names]
test_y = y


X_constant = sm.add_constant(test_x)
sm_model = sm.OLS(test_y, X_constant).fit()
print(sm_model.summary())
```

Summary
---

### Significance:
Since all P-values are 0.000, we can be very confident that these relationships aren't just due to random chance.

### Reliability:
The Standard Errors are generally very low (especially for study hours and attendance), meaning these estimates are precise.

### Top Performance Drivers:
Study Hours and Coaching are the most influential factors for increasing a student's score.
