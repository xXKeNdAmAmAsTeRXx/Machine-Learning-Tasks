# Best Model fitted with all data

## Data

```python
import pandas as pd

data = pd.read_csv('new_Exam_Score_Prediction.csv')

y = data['exam_score']
X = data.drop('exam_score', axis=1)
X
```

```python
import pickle
import os

with open(os.path.join('pipes', 'Linear.pkl'), 'rb') as f:
    Pipe = pickle.load(f)

Pipe.fit(X,y)
```

# Making a prediction

## Observation

```python
MyParams = pd.DataFrame(
    {
        'study_hours': [7],
        'class_attendance': [0.9],  # (0,1)
        'facility_rating': [2],     # 0/1/2
        'gender': [1],              # Male: 1, Female: 0
        'coaching': [0],            # 0/1
        'self_study': [1]           # 0/1
    }
)

MyParams.T
```

## Prediction

```python
import numpy as np
pred = Pipe.predict(MyParams)

# As exam score lower than 0 and higher than 100 does not make any sense
pred =np.clip(pred,0,100)
pred
```
