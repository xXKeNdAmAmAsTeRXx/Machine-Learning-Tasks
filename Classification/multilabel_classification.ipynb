```python
import pandas as pd

from sklearn.datasets import make_multilabel_classification
from sklearn.model_selection import train_test_split

```

# DataSet

```python
data = make_multilabel_classification(n_samples=2000, n_classes=5, n_features=20, n_labels=2, random_state=42)


data_df = pd.DataFrame(data[0])

data_df
```

```python
## Data Correlation
```

```python
import matplotlib.pyplot as plt

names = data_df.columns[:20]
corr_mtx = (data_df[names]).corr(method='pearson')
plt.matshow(corr_mtx, cmap="grey")
plt.colorbar()

plt.xticks(range(len(names)), names, rotation=90)
plt.yticks(range(len(names)), names)
plt.show()

corr_mtx
```

## Data Split

```python
X = data[0]
y = data[1]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3)
```

# MultiLabels Model (Wrappers)

```python
from sklearn.multioutput import MultiOutputClassifier
```

## Metric Function

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import (
    accuracy_score,
    multilabel_confusion_matrix,
    roc_auc_score,
    classification_report, ConfusionMatrixDisplay
)

def metrics_multi_labels(y_true: np.ndarray, y_pred_classes: np.ndarray, y_pred_scores: np.ndarray, method: str) -> pd.DataFrame:


    classes = np.unique(y_true)
    cr = classification_report(y_true, y_pred_classes, zero_division=1.0)

    acc = accuracy_score(y_true, y_pred_classes)
    print(cr)
    print(f"Accuracy Score: {acc}")

    return cr



model_performence = {}
```

## Logistic Regression

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LogisticRegression
import optuna

optuna.logging.set_verbosity(optuna.logging.ERROR)


def objective_lr(trial):
    C = trial.suggest_float('C', 1e-3, 1e2, log=True)
    solver = trial.suggest_categorical('solver', ['lbfgs', 'liblinear', 'saga'])
    max_iter = trial.suggest_int('max_iter', 100,300)
    n_jobs = trial.suggest_int('n_jobs', 1, 10)

    clf = MultiOutputClassifier(LogisticRegression(C=C, solver=solver, max_iter=max_iter, random_state=42), n_jobs=n_jobs)
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    score = cross_val_score(clf, X_train, y_train, cv=cv, scoring='roc_auc_ovr').mean()

    return score



study = optuna.create_study(direction='maximize')
study.optimize(objective_lr, n_trials=100)

study.best_params
```

### Model fit

```python
lr_model = MultiOutputClassifier(LogisticRegression(C=study.best_params['C'], solver=study.best_params['solver'], max_iter=study.best_params['max_iter'], random_state=42), n_jobs=study.best_params['n_jobs'])

#lr_model = MultiOutputClassifier(LogisticRegression())
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
y_prob = lr_model.predict_proba(X_test)

model_performence["LogisticRegression"] = metrics_multi_labels(y_test, y_pred, y_prob, "LogisticRegression")
```

## KNN


```python
from sklearn.neighbors import KNeighborsClassifier

def objective_KNN(trail):
    n_neighbors = trail.suggest_int('n_neighbors', 1,30)
    weights = trail.suggest_categorical('weights', ['uniform', 'distance'])
    metric = trail.suggest_categorical('metric', ['euclidean', 'manhattan', 'minkowski', 'chebyshev'])

    clf = MultiOutputClassifier(KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights, metric=metric))
    clf.fit(X_train, y_train)
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    score = cross_val_score(clf, X_train, y_train, cv=cv, scoring='roc_auc_ovr').mean()

    return score

knn_study = optuna.create_study(direction="maximize")
knn_study.optimize(objective_KNN)
knn_study.optimize(objective_KNN, n_trials=100)

knn_study.best_params
```

### Model fit

```python
knn_model = MultiOutputClassifier(KNeighborsClassifier(n_neighbors=knn_study.best_params['n_neighbors'], weights=knn_study.best_params['weights'], metric=knn_study.best_params['metric']))
knn_model.fit(X_train, y_train)
```

```python
y_pred = knn_model.predict(X_test)
y_prob = knn_model.predict_proba(X_test)

model_performence["KNN"] = metrics_multi_labels(y_test, y_pred, y_prob, "KNN")
```

## LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

def object_LDA(trial):
    solver = trial.suggest_categorical('solver', ['lsqr', 'eigen'])
    shrinkage = trial.suggest_float('shrinkage', 0, 1, log=False)

    clf = MultiOutputClassifier(LinearDiscriminantAnalysis(solver=solver, shrinkage=shrinkage))
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    score = cross_val_score(clf, X_train, y_train, cv=cv, scoring='roc_auc_ovr').mean()
    return score


LDA_study = optuna.create_study(direction="maximize")
LDA_study.optimize(object_LDA, n_trials=100)

LDA_study.best_params
```

### Model fit

```python
LDA_model = MultiOutputClassifier(LinearDiscriminantAnalysis(solver=LDA_study.best_params['solver'], shrinkage=LDA_study.best_params['shrinkage']))
LDA_model.fit(X_train, y_train)
```

```python
y_pred = LDA_model.predict(X_test)
y_prob = LDA_model.predict_proba(X_test)

model_performence["LDA"] = metrics_multi_labels(y_test, y_pred, y_prob, "LDA")
```
