# LOOCV Error Estimate

```python
from random import sample

from ISLP import load_data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data

```python
data = load_data('Weekly')

data
```

```python
x = data[['Lag1', 'Lag2']].values
y = np.array([1 if d=="Up" else 0 for d in data['Direction']])
```

## LOOCV Loop


```python
def LOOCV_acc_score(x,y, idx):
    x_test,y_test = x[idx].reshape(1, -1), y[idx]

    mask = np.ones(len(x), dtype=bool)
    mask[idx] = False

    x_train, y_train = x[mask][:],y[mask][:]
    model = LogisticRegression()
    model.fit(x_train, y_train)

    pred = model.predict(x_test)

    if pred == y_test:
        return 1
    else:
        return 0
```

```python
samples_num = len(x)
num_correct = 0
for i in range(len(y)):
    num_correct += LOOCV_acc_score(x,y, i)

LOOCV_error = (samples_num - num_correct )/ samples_num
print(f"{LOOCV_error=}")
```
